Una clase es una entidad que define una serie de elementos que determinan un estado (datos) y un comportamiento (operaciones sobre los datos que modifican su estado).

Por su parte, un objeto es una concrecin o instancia de una clase.

Seguro que si te digo que te imagines un coche, en tu mente comienzas a visualizar la carrocera, el color, las ruedas, el volante, si es disel o gasolina, el color de la tapicera, si es manual o automtico, si acelera o va marcha atrs, etc.

Pues todo lo que acabo de describir viene a ser una clase y cada uno de los de coches que has imaginado, seran objetos de dicha clase.





```python
class Coche:
    ruedas = 4 #Atributo de clase
    
    def __init__(self, color, aceleracion):
        self.color = color
        self.aceleracion = aceleracion
        self.velocidad = 0
    def acelera(self):
        self.velocidad = self.velocidad + self.aceleracion
    def frena(self): #Mtodo
        v=self.velocidad - self.aceleracion
        if v<0:
            v=0
        self.velocidad = v
            
```

El esquema anterior define la clase Coche (es una versin muy, muy simplificada de lo que es un coche,  pero nos sirve de ejemplo). Dicha clase establece una serie datos, como ruedas, color, aceleracin o velocidad y las operaciones acelera() y frena().

Cuando se crea una variable de tipo Coche, realmente se est instanciando un objeto de dicha clase. En el siguiente ejemplo se crean dos objetos de tipo Coche:

```python
c1 = Coche('rojo', 20)
print(c1.color)
```

```python
print(c1.ruedas)
4
```

# Mtodos
Como te explicaba al comienzo de esta seccin, los mtodos son las funciones que se definen dentro de una clase y que, por consiguiente, pueden ser referenciadas por los objetos de dicha clase. Sin embargo, realmente los mtodos son algo ms.

Si te has fijado bien, pero bien de verdad, habrs observado que las funciones acelera() y frena() definen un parmetro self.

No obstante, cuando se usan dichas funciones no se pasa ningn argumento. Qu est pasando? Pues que acelera() est siendo utilizada como un mtodo por los objetos de la clase Coche, de tal manera que cuando un objeto referencia a dicha funcin, realmente pasa su propia referencia como primer parmetro de la funcin.

 NOTA: Por convencin, se utiliza la palabra self para referenciar a la instancia actual en los mtodos de una clase.

Sabiendo esto, podemos entender, por ejemplo, por qu todos los objetos de tipo Coche pueden referenciar a los atributos de datos velocidad o color. Son inicializados para cada objeto en el mtodo __init__().

# Atributos de clase y atributos de instancia
Una clase puede definir dos tipos diferentes de atributos de datos: atributos de clase y atributos de instancia.

Los atributos de clase son atributos compartidos por todas las instancias de esa clase.
Los atributos de instancia, por el contrario, son nicos para cada uno de los objetos pertenecientes a dicha clase.
En el ejemplo de la clase Coche, ruedas se ha definido como un atributo de clase, mientras que color, aceleracion y velocidad son atributos de instancia.

Para referenciar a un atributo de clase se utiliza, generalmente, el nombre de la clase. Al modificar un atributo de este tipo, los cambios se vern reflejados en todas y cada una las instancias.

```python
c1 = Coche('rojo', 20)
c2 = Coche('azul', 20)
```

```python
print(c1.color)
print(c2.color)
```

```python
print(c1.ruedas)  # Atributo de clase
print(c2.ruedas)  # Atributo de clase
```

```python
Coche.ruedas = 6  # Atributo de clase
```

```python
print(c1.ruedas)  # Atributo de clase
print(c2.ruedas)  # Atributo de clase
```

Si un objeto modifica un atributo de clase, lo que realmente hace es crear un atributo de instancia con el mismo nombre que el atributo de clase.

```python
c1 = Coche('rojo', 20)
c2 = Coche('azul', 20)
```

```python
print(c1.color)
print(c2.color)
```

```python
c1.ruedas = 6  # Crea el atributo de instancia ruedas
```

```python
print(c1.ruedas)
print(c2.ruedas)
```

```python
print(Coche.ruedas)
```

# Herencia en Python
En programacin orientada a objetos, la herencia es la capacidad de reutilizar una clase extendiendo su funcionalidad. Una clase que hereda de otra puede aadir nuevos atributos, ocultarlos, aadir nuevos mtodos o redefinirlos.

En Python, podemos indicar que una clase hereda de otra de la siguiente manera:

```python
class CocheVolador(Coche):

    ruedas = 6

    def __init__(self, color, aceleracion, esta_volando=False):
        super().__init__(color, aceleracion)
        self.esta_volando = esta_volando

    def vuela(self):
        self.esta_volando = True

    def aterriza(self):
        self.esta_volando = False
```

Como puedes observar, la clase CocheVolador hereda de la clase Coche. En Python, el nombre de la clase padre se indica entre parntesis a continuacin del nombre de la clase hija.

La clase CocheVolador redefine el atributo de clase ruedas, estableciendo su valor a 6 e implementa dos mtodos nuevos: vuela() y aterriza().

Fjate ahora en la primera lnea del mtodo __init__(). En ella aparece la funcin super(). Esta funcin devuelve un objeto temporal de la superclase que permite invocar a los mtodos definidos en la misma. Lo que est ocurriendo es que se est redefiniendo el mtodo __init__() de la clase hija usando la funcionalidad del mtodo de la clase padre. Como la clase Coche es la que define los atributos color y aceleracion, estos se pasan al constructor de la clase padre y, a continuacin, se crea el atributo de instancia esta_volando solo para objetos de la clase CocheVolador.

Al utilizar la herencia, todos los atributos (atributos de datos y mtodos) de la clase padre tambin pueden ser referenciados por objetos de las clases hijas. Al revs no ocurre lo mismo.

Veamos todo esto con un ejemplo:

```python
c = Coche('azul', 10)
cv1 = CocheVolador('rojo', 60)
```

```python
print(cv1.color)
print(cv1.esta_volando)
```

```python
cv1.acelera()
```

# Las funciones isinstance() e issubclass()
Como ya vimos en otros tutoriales, la funcin incorporada type() devuelve el tipo o la clase a la que pertenece un objeto. En nuestro caso, si ejecutamos type() pasando como argumento un objeto de clase Coche o un objeto de clase CocheVolador obtendremos lo siguiente:

```python
c = Coche('rojo', 20)
type(c)
```

```python
cv = CocheVolador('azul', 60)
type(cv)
```

Sin embargo, Python incorpora otras dos funciones que pueden ser de utilidad cuando se quiere conocer el tipo de una clase. Son: isinstance() e issubclass().

isinstance(objeto, clase) devuelve True si objeto es de la clase clase o de una de sus clases hijas. Por tanto, un objeto de la clase CocheVolador es instancia de CocheVolador pero tambin lo es de Coche. Sin embargo, un objeto de la clase Coche nunca ser instancia de la clase CocheVolador.
issubclass(clase, claseinfo) comprueba la herencia de clases. Devuelve True en caso de que clase sea una subclase de claseinfo, False en caso contrario. claseinfo puede ser una clase o una tupla de clases.

```python
c = Coche('rojo', 20)
cv = CocheVolador('azul', 60)
```

```python
isinstance(c, Coche)
```

```python
isinstance(cv, Coche)
```

```python
isinstance(c, CocheVolador)
```

```python
isinstance(cv, CocheVolador)
```

```python
issubclass(CocheVolador, Coche)
```

```python
issubclass(Coche, CocheVolador)
```
